```python
def impute_numerical_mean(data):
    
    imp_mean = IterativeImputer(max_iter=10, random_state=0)
    d = imp_mean.fit_transform(data)
    return d

def impute_cat_mode(data):
    
    imp_mean = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
    d = imp_mean.fit_transform(data)
    return d

# return missing variable that contain missing values
def get_missing_cols(df):
    data = df.columns[df.isna().any()].tolist()
    return data

# imputing values
def impute_columns(df):
    
    missing_col = get_missing_cols(df)
    
    for i in missing_col:
        if df[i].dtype == 'O':
            df[i] = impute_cat_mode(df[i].values.reshape(-1, 1))
        else:
            df[i] = impute_numerical_mean(df[i].values.reshape(-1, 1))

    
def get_data_impute(df):
    
    # return df with imputed missing values
    df_temp_imputed = df.copy()
    impute_columns(df_temp_imputed)
    return df_temp_imputed
```

```python
# different imputers

def impute_numerical_mean(data):
    
    imp_mean = SimpleImputer(missing_values=np.nan, strategy='mean')
    d = imp_mean.fit_transform(data)
    return d

def impute_cat_mode(data):
    
    imp_mean = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
    d = imp_mean.fit_transform(data)
    return d

# return missing variable that contain missing values
def get_missing_cols(df):
    data = df.columns[df.isna().any()].tolist()
    return data

# imputing values
def impute_columns(df):
    
    missing_col = get_missing_cols(df)
    
    for i in missing_col:
        if df[i].dtype == 'O':
            df[i] = impute_cat_mode(df[i].values.reshape(-1, 1))
        else:
            df[i] = impute_numerical_mean(df[i].values.reshape(-1, 1))

    
def get_data_impute(df):
    
    # return df with imputed missing values
    df_temp_imputed = df.copy()
    impute_columns(df_temp_imputed)
    
    return df_temp_imputed

def get_data_drop(df):
    # return df without NaN's
    df_temp_drop_na = df.copy()
    
    df_temp_drop_na.dropna(inplace=True)
    df_temp_drop_na.reset_index(inplace=True)
    df_temp_drop_na.drop('index', axis=1, inplace=True)
    
    return df_temp_drop_na
```

```python
def get_encoder_inst(feature_col):
    """
    returns: an instance of sklearn OneHotEncoder fit against a (training) column feature;
    such instance is saved and can then be loaded to transform unseen data
    """
    assert isinstance(feature_col, pd.Series)
    feature_vec = feature_col.sort_values().values.reshape(-1, 1)
    enc = OneHotEncoder(handle_unknown='ignore')
    enc.fit(feature_vec) 
    
    joblib.dump(enc, f'./encoders/one_hot_encoder_{feature_col.name}.pkl')
    return enc
def get_one_hot_enc(feature_col, enc):
    """
    maps an unseen column feature using one-hot-encoding previously fit against training data 
    returns: a pd.DataFrame of newly one-hot-encoded feature
    """
    assert isinstance(feature_col, pd.Series)
    assert isinstance(enc, OneHotEncoder)
    unseen_vec = feature_col.values.reshape(-1, 1)
    encoded_vec = enc.transform(unseen_vec).toarray()
    encoded_df = pd.DataFrame(encoded_vec)
    return encoded_df
```

```python
def grid_search_(name, model, X, y):
    
    # Check if there are existing parameters, else execute the grid search for the chosen model
    if os.path.isfile(f'./model_parameters/best_params_{name}.pkl'):
        params = joblib.load(f'./model_parameters/best_params_{name}.pkl')
        return params
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
    
    # 
    parameters = {"alpha": [1, 10, 100, 290, 500],
              "fit_intercept": [True, False],
              "solver": ['svd', 'cholesky', 'lsqr', 'sparse_cg', 'sag', 'saga'], 
             }
    
    if name == 'RF':
        # Randomforests parameters
        n_estimators = [int(x) for x in np.linspace(start = 200, stop = 2000, num = 10)]
        max_features = ['auto', 'sqrt']
        max_depth = [int(x) for x in np.linspace(10, 110, num = 11)]
        max_depth.append(None)
        min_samples_split = [2, 5, 10]
        min_samples_leaf = [el for el in range(1,10)]
        bootstrap = [True, False]
        random_grid = {'n_estimators': n_estimators,
                       'max_features': max_features,
                       'max_depth': max_depth,
                       'min_samples_split': min_samples_split,
                       'min_samples_leaf': min_samples_leaf,
                       'bootstrap': bootstrap}
    elif name == 'XGB':
        # Xgboost parameters
        random_grid = {
            'min_child_weight': [1, 5, 10],
            'gamma': [0.5, 1, 1.5, 2, 5],
            'subsample': [0.6, 0.8, 1.0],
            'colsample_bytree': [0.6, 0.8, 1.0],
            'max_depth': [el for el in range(1,20)]
            }
    
    if name == 'XGB':
        grid_model = XGBRegressor(learning_rate =0.1, n_estimators=1000)
    else:
        grid_model = model
    grid_search = RandomizedSearchCV(estimator = grid_model, param_distributions = random_grid, n_iter = 100, cv = 3, verbose=2, random_state=42, n_jobs = -1)
    grid_search.fit(X_train, y_train)
    
    best_params = grid_search.best_params_
    joblib.dump(best_params, f'./model_parameters/best_params_{name}.pkl')
    
    
    return best_params
```

```python
# compare multiple models
from xgboost import XGBRegressor
def model_comparison(X, y):
    
    results = []
    models_out = []
    
    models = [('RF', RandomForestRegressor()), ('XGB', XGBRegressor())]
   
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

    for name, algo in models:
        
        print(f'Fitting {name}')
        
        #get params
        params = grid_search_(name,algo, X, y)
        
        if name == 'RF':
            model = RandomForestRegressor(**params)
            model.fit(X_train, y_train)
        elif name == 'XGB':
            model = XGBRegressor(learning_rate =0.1, n_estimators=1000, **params)
            model.fit(X_train.values, y_train.values)
            
        # model predictions
        y_hat = model.predict(X_test.values)
        
        # evaluate r2 and rmse
        r2 = r2_score(y_test, y_hat)
        mse = mean_squared_error(y_test, y_hat, squared=False)

        results.append((model, name, r2, mse))
        
        # append trained models with labels
        models_out.append((name, model))
        
    return results, models_out, X_train, X_test, y_train, y_test, y_hat
```

```python
def save_model(model, name):
    joblib.dump(model, f"{name}.joblib")
```

```python
def train_model(X, y):
    
    # Train test data split 90 / 10
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
    
    #out of the box model
    model = RandomForestRegressor(random_state = 42)
    model.fit(X_train, y_train)
    y_hat_baseline = model.predict(X_test)
    baseline_score_r2, baseline_score_rmse = evaluate(model, y_hat_baseline, y_test)
    
    # optimised model
    get_params = joblib.load('best_params_rf_2.pkl')
    model_optimal  = RandomForestRegressor(**get_params, random_state = 42)
    model_optimal.fit(X_train, y_train)
    y_hat = model_optimal.predict(X_test)
    optimised_score_r2, optimised_score_r2_rmse  = evaluate(model, y_hat, y_test)
    
    print('Improvement of R2 metric by {:0.2f}%.'.format( 100 * (optimised_score_r2 - baseline_score_r2) / baseline_score_r2))
    print('Improvement of RMSE metric by {:0.2f}%.'.format( 100 * (optimised_score_r2_rmse - baseline_score_rmse) / baseline_score_rmse))
    
    return model, X_train, X_test, y_train, y_test, y_hat
    
def evaluate(model, y_hat, y_test):
    #print("r2 :", r2_score(y_test, y_hat))
    rms = mean_squared_error(y_test, y_hat, squared=False)
    r2 = r2_score(y_test, y_hat)
    return r2, rms



def grid_search(X, y):
    
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
    
    # Number of trees in random forest
    n_estimators = [int(x) for x in np.linspace(start = 200, stop = 2000, num = 10)]
    # Number of features to consider at every split
    max_features = ['auto', 'sqrt']
    # Maximum number of levels in tree
    max_depth = [int(x) for x in np.linspace(10, 110, num = 11)]
    max_depth.append(None)
    # Minimum number of samples required to split a node
    min_samples_split = [2, 5, 10]
    # Minimum number of samples required at each leaf node
    min_samples_leaf = [el for el in range(1,10)]
    # Method of selecting samples for training each tree
    bootstrap = [True, False]
    # Create the random grid
    random_grid = {'n_estimators': n_estimators,
                   'max_features': max_features,
                   'max_depth': max_depth,
                   'min_samples_split': min_samples_split,
                   'min_samples_leaf': min_samples_leaf,
                   'bootstrap': bootstrap}
    
    rf = RandomForestRegressor()
    rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid, n_iter = 300, cv = 3, verbose=2, random_state=42, n_jobs = -1)
    rf_random.fit(X_train, y_train)
    
    best_params = rf_random.best_params_
    
    return best_params
```

```python
# run a grid search and save the best parameters
def optimise_model(X, y):
    params = grid_search(X, y)
    return params
    #joblib.dump(params, 'best_params_rf.pkl')
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.layers import Dropout
from tensorflow.keras.callbacks import EarlyStopping

def train_NN_model(X, y):
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
    
    model = Sequential()
    model.add(Dense(7, input_dim=7, kernel_initializer='normal', activation='relu'))
    model.add(Dense(1024, activation='relu'))
    model.add(Dense(512, activation='relu'))
    model.add(Dense(256, kernel_initializer='normal', activation='relu'))
    model.add(Dense(128, kernel_initializer='normal', activation='relu'))
    model.add(Dense(64, kernel_initializer='normal', activation='relu'))
    model.add(Dense(1, kernel_initializer='normal'))
    # Compile model
    model.compile(loss='mean_squared_error', optimizer='adam')

    model.fit(x=X_train,y=y_train.values,
              validation_data=(X_test,y_test.values),
              batch_size=256,epochs=500)
    
    y_hat = model.predict(X_test)
    
    return model, X_train, X_test, y_train, y_test, y_hat 
```
